```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score as r2
from sklearn.metrics import mean_squared_error as mse
```

```python
data1=pd.read_excel('es4029414_si_001.xlsx')
data1
```

```python
fit=list(data1.iloc[:,1])
pred=list(data1.iloc[:,2])
torv=list(data1.iloc[:,3])
train_fit=[]
trian_pre=[]
test_fit=[]
test_pre=[]
n=0
m=0
for i in range(len(torv)):
    if torv[i]=='T':
        n+=1
        train_fit.append(fit[i])
        trian_pre.append(pred[i])
    elif torv[i]=='V':
        m+=1
        test_fit.append(fit[i])
        test_pre.append(pred[i])
    else:
        print(i)
```

```python
print(n,m)
```

```python
plt.figure(figsize=(10,7))

font2 = {'family' : 'Times New Roman',
'weight' : 'bold',
'size' : 45,
}
plt.scatter(trian_pre,train_fit,alpha=.5, label='train')
plt.scatter(test_pre,test_fit, alpha=.5, label='test1')
#plt.scatter(newX3[:, 0], newX3[:, 1], alpha=.5, label='test2')
plt.legend()
plt.tick_params(labelsize=15)
plt.legend(prop={'family' : 'Times NewRoman', 'size'   : 20})
plt.xlabel('pre',fontsize=16)
plt.ylabel('real',fontsize=16)
#plt.show()
```
