```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_curve,auc
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score as ACC
from sklearn.metrics import matthews_corrcoef as MCC
from sklearn.metrics import roc_auc_score as AUC
from sklearn.metrics import f1_score as F1
from sklearn.model_selection import train_test_split
def auc_roc(y_test1,pre):
    pre=[p for p in pre[:,1]]
    #print(pre)
    fpr,tpr,threshold = roc_curve(y_test1,pre)###计算真正率和假正率
    roc_auc = auc(fpr,tpr)
    #print('roc_auc:%2f'%(roc_auc))
    return roc_auc
def analyze(pre,real):
    AUC_test=auc_roc(real,pre)
    pre=np.argmax(pre,axis=1)
    #print(pre)
    m=confusion_matrix(pre,real, labels=[0,1], sample_weight=None)
    print(m)
    tp, fp, fn, tn = m.ravel()
    TPR=tp/(tp+fn)
    TNR=tn/(tn+fp)
    ACC_test=ACC(real,pre)
    #AUC_test=auc_roc(real,pre)
    MCC_test=MCC(real,pre)
    F1_test=F1(real,pre)
    print('敏感度TPR：%2f，特异度TNR：%2f，ACC：%2f，AUC：%2f,MCC：%2f,F1：%2f'%(TPR,TNR,ACC_test,AUC_test,MCC_test,F1_test))
    return ACC_test
import pickle
def save_rfmodel(name,model):
    model_name=name+'.pkl'
    with open(model_name, 'ab') as f:
        pickle.dump(model, f)
def rfmodel_pre(model_name,x,y):
    with open(model_name+'.pkl','rb') as md:
        model=pickle.load(md)
    pre=prediction(model,x,y)
    return pre
def prediction(model,x,y):
    pre_x=model.predict_proba(x)
    analyze(pre_x,y)
    return pre_x
```

```python
import seaborn as sb
from rdkit.Chem import MACCSkeys
from rdkit import Chem
from rdkit import DataStructs
import matplotlib.pyplot as plt

def get_fps(smiles):
    all_mols=[]
    for smi in list(smiles):
        m=Chem.MolFromSmiles(smi)
        all_mols.append(m)
    fps=[Chem.RDKFingerprint(mol) for mol in all_mols]
    return fps
def similar_fenbu(test,train):
    test_fps=get_fps(test)
    train_fps=get_fps(train)
    similars=[]
    for i in range(len(test_fps)):
        fp1=test_fps[i]
        similar=[]
        for y in range(len(train_fps)):
            fp2=train_fps[y]
            sim=(DataStructs.FingerprintSimilarity(fp1,fp2))
            #if sim==1.0:
            #print(list(test_datas.iloc[:,0])[i],list(train_datas.iloc[:,0])[y])
            similar.append(sim)
        similars.append(max(similar))
    print(max(similars))
    similars_int=[]
    for sim in similars:
        similars_int.append(round(sim,1))
    similar_result=pd.DataFrame()
    fenbu_name='similarity'
    similar_result[fenbu_name]=similars
    similar_result[fenbu_name+'int']=similars_int
    #sb.regplot(data = similar_result,y = fenbu_name, x=[1]*len(similars), fit_reg = False,
    #      x_jitter =0.2, scatter_kws = {'alpha' : 1/3,'s':10})
    #sb.swarmplot(y='similar',data=similar_result)#y=list(similar_result['similar']))
    print(np.array(list(similar_result['similarity'])).mean())
    #plt.show()
    return similar_result
```

# 基学习器的多样性

```python
from abc import ABC, abstractmethod
import numpy as np



#def get_measure(self, predictors, X, y):
    #pass

#PCDM
def get_measure1(predictors, X, y):
    """
    Measures the Percentage of Correct Diversity.
    :param predictors: <list> Decision Trees
    :param X: <numpy ndaray> Feature vectors
    :param y: <numpy array> Target feature
    :return: <float>
    """
    tally = 0
    n_instances = X.shape[0]
    for i in range(n_instances):
        instance, target = X[i], y[i]
        n_corrects = 0
        for p in predictors:
            prediction = p.predict(instance)
            if prediction == target:
                n_corrects += 1
        if 0.1 * len(predictors) <= n_corrects <= 0.9 * len(predictors):
            tally += 1
    diversity = tally / n_instances
    return diversity

#Q
def get_measure2( predictors, X, y):
    """
    Measures the QStatistic Diversity.
    :param predictors: <list> Decision Trees
    :param X: <numpy ndaray> Feature vectors
    :param y: <numpy array> Target feature
    :return: <float>
    """
    n_instances = X.shape[0]
    n_predictors = len(predictors)
    q_total = 0
    for i in range(0, n_predictors-1):
        for j in range(i+1, n_predictors):
            n = np.zeros((2, 2))
            for k in range(n_instances):
                i_pred = predictors[i].predict(X[k])
                j_pred = predictors[j].predict(X[k])
                true_y = y[k]
                if i_pred == true_y:
                    if j_pred == true_y:
                        n[1][1] += 1
                    else:
                        n[1][0] += 1
                else:
                    if j_pred == true_y:
                        n[0][1] += 1
                    else:
                        n[0][0] += 1

            # Adding a one value to the variables which are zeros
            for k in range(2):
                for l in range(2):
                    if n[k][l] == 0:
                        n[k][l] += 1
            same = n[1][1] * n[0][0]
            diff = n[1][0] * n[0][1]
            q_ij = (same - diff) / (same + diff)
            q_total += q_ij

    q_av = 2 * q_total / (n_predictors * (n_predictors - 1))
    return q_av
```

```python
import joblib
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
def maxabs(a, axis=None):
    """Return slice of a, keeping only those values that are furthest away
    from 0 along axis"""
    maxa = a.max(axis=axis)
    mina = a.min(axis=axis)
    p = abs(maxa) > abs(mina) # bool, or indices where +ve values win
    n = abs(mina) > abs(maxa) # bool, or indices where -ve values win
    if axis == None:
        if p: return maxa
        else: return mina
    shape = list(a.shape)
    shape.pop(axis)
    out = np.zeros(shape, dtype=a.dtype)
    out[p] = maxa[p]
    out[n] = mina[n]
    return out
```

```python
def test(d):
    f=[]
    for i in range(d.shape[0]):
        if d[i][0]>0.4:
            continue
        else:
            if d[i][1]>0.2:
                continue
            elif d[i][2]>0.4:
                continue
            else:
                f.append(d[i])
    return np.array(f)
```
